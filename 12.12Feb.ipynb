```python
# Q5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example
#Answer : Custom exceptions in Python are user-defined exceptions that extend the functionality of the built-in exceptions. Custom 
        # exceptions in Python are used to create our own exception classes to handle specific error conditions 
        #In Python there are many  predefined exception but not all, if we want to raise any exception according to our spcefic condition 
        #then we need custom exceptions
#Example
def agevalidate(age):
    if age >= 18:
        print("he/she is eligible for vote!!")
    else :
        print("he/she is not eligible for vote!!")
agevalidate(19)
agevalidate(17)
agevalidate(-2)  #in this case we need custom exception handling because age can't be negative, we can handle it with if else but the best 
                # practice is to use custom exception handling
    
```

```python
# Q6. Create custom exception class. Use this class to handle an exception.
#Answer 
class WithdrawalError(Exception):

    def __init__(self):
        print("Insufficient balance")

def Withdrawal(balance, amount):
    if amount > balance :
        raise WithdrawalError()
    else:
        balance -= amount
        print("Withdrawal Successfull!!",balance)

try:
    Withdrawal(1000,1500)
except Exception as e:
    print(e)
```
